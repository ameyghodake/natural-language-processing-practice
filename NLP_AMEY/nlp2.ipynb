# Train own model

```python
import gensim
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
msg= pd.read_csv('C:/Users/Amey/Documents/spam.csv',encoding='latin-1')
msg.head()
```

```python
msg = msg.drop(labels= ['Unnamed: 2','Unnamed: 3','Unnamed: 4'], axis =1)
msg.head()
```

```python
msg.columns = ['label', 'text']
msg.head()
```

```python
msg['text_clean']= msg['text'].apply(lambda x: gensim.utils.simple_preprocess(x))
msg.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split( msg['text_clean'], 
                                                    msg['label'], 
                                                   test_size= 0.2)
```

```python
w2v_model= gensim.models.Word2Vec(x_train,
                                vector_size= 100,
                                window= 5,
                                min_count=2)
```

```python
pd.set_option('display.max_colwidth', 100)
msg['text_clean'] = msg['text'].apply(lambda x: gensim.utils.simple_preprocess(x))
X_train, X_test, y_train, y_test = train_test_split(msg['text_clean'],
                                                    msg['label'], test_size=0.2)

w2v_model = gensim.models.Word2Vec(X_train,
                                   window=5,
                                   min_count=2)
```

```python
w2v_model.wv['king']
```

```python
w2v_model.wv.most_similar('king')
```

# Generate Aggregate sentence vectors

```python
w2v_vect = np.array([np.array([w2v_model.wv[i] for i in ls if i in w2v_model.wv.index_to_key] )
                     for ls in x_test])
w2v_vect
```

```python
for i,v in enumerate(w2v_vect):
    print(len(x_test.iloc[i]), len(v))
```


# Need to take average of word vectors

No of Features will be same. ie 100

```python
w2v_vect_avg = []
for vect in w2v_vect:
    w2v_vect_avg.append(vect.mean(axis= 0))
else:
    w2v_vect_avg.append(np.zeros(100))
```

```python
for i,v in enumerate(w2v_vect_avg):
    print(len(x_test.iloc[i]), len(v))
```
