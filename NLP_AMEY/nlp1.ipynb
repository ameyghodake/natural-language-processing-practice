```python
import nltk
```

```python
import pandas as pd
import re
import string
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('stopwords')
```

```python
stopwords= nltk.corpus.stopwords.words('english')
```

```python
msg= pd.read_csv('C:/Users/Amey/Documents/spam.csv',encoding='latin-1');
```

```python
msg.head()
```

```python
msg= msg.drop(labels=['Unnamed: 2','Unnamed: 3','Unnamed: 4'], axis=1)
```

```python
msg.head()
```

```python
msg.columns= ['label','text']
msg.head()
```

```python
def clean_text(text):
    text= ' '.join([word.lower() for word in text if word not in string.punctuation])
    tokens= re.split('\W+',text)
    text= [word for word in tokens if word not in stopwords]
    return text
```

```python
tfidf_vect= TfidfVectorizer(analyzer= clean_text)
x_tfidf= tfidf_vect.fit_transform(msg['text'])
```

```python
x_features= pd.DataFrame(x_tfidf.toarray())
x_features
```

```python
def remove_punct(text):
    text= ' '.join([char for char in text if char not in string.punctuation])
    return text
```

# #Build the model

```python
from sklearn.ensemble  import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_features, msg['label'], test_size=0.2)
```

```python
rf= RandomForestClassifier()

rf_model= rf.fit(x_train, y_train)
```

```python
y_pred = rf_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
precision = precision_score(y_test, y_pred, pos_label='spam')
recall =   recall_score(y_test, y_pred, pos_label='spam')
print('precision:{} / recall: {}'.format(round(precision,3), round(recall,3)))
```
